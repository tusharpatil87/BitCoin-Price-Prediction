There are many factors which are affecting the price of Bitcoin, that's the reason we are looking to predict the future price of the Bitcoin with the help of Open source libraries like Python, Facebook Prophet.


```python
import numpy as np
import pandas as pd
from fbprophet import Prophet
```

```python
path = r"E:\Skills\Machine_Larning\DATA_SCIENCE_PROJECT\2021\BitCoin\BTC-USD.csv"
```

```python
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df = df[["Date", "Close"]]
```

```python
df.head()
```

```python
df.columns = ["ds", "y"]
```

```python
df.head()
```

```python
import fbprophet
print(fbprophet.__version__)
```

```python
pro = Prophet()
```

```python
pro.fit(df)
```

```python
future = pro.make_future_dataframe(periods=365)
print(future)
```

```python
df.shape
```

```python
forecast = pro.predict(future)
```

```python
forecast
```

```python
forecast[["ds", "yhat", "yhat_lower", "yhat_upper"]].tail(200)
```

```python
from fbprophet.plot import plot
pro.plot(forecast, figsize=(10, 10))
```
